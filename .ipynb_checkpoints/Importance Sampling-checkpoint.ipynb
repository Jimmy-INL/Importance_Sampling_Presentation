```python
import numpy as np
import seaborn as sns
from tqdm import trange
from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.hist(np.random.normal(0, 100, 10**5), bins=50)
plt.show()

original_pdf = np.random.normal(0, 0.01, 10**5)
optimised_pdf = np.random.normal(0, 0.06, 10**5)
weights = original_pdf/optimised_pdf
plt.hist(weights, bins=50)
plt.show()
plt.hist(original_pdf, bins=50)
plt.show()
```

```python
def spike(x, center=8):
    y = -np.power(x, 2)-center
    #y[y < 0] = 0
    return y

def importance_sampling(nsamples=10, nsimulations=10000):
    # Basically this: https://github.com/kevinzakka/monte-carlo/blob/master/Monte%20Carlo%20and%20Importance%20Sampling.ipynb
    original_pdf = norm(0, 0.01)
    optimised_pdf = norm(0, 0.02)
    
    for i in range(nsimulations):
        a = np.random.normal(0,0.01,nsamples)
        
        weights = original_pdf(a)/optimised_pdf(a)
        
        
    return
        
```
